```python
test_df = BC_df.head()
```

```python
testout_df = pd.DataFrame()
testout_df
```

```python
# break the barcode string
BC_list = []
for _, row in test_df.iterrows():
    Barcode = row['Barcode']
    BC = 'T'+ Barcode + 'A'
    sp1 = BC[0:3]
    sp2 = BC[3:6]
    sp3 = BC[6:8]
    sp4 = BC[8:11]
    sp5 = BC[11:14]
    NewBC = sp1+'T'+ sp2 + 'A' +sp3 + 'T'+sp4+'A'+sp5
    BC_list.append(NewBC)
    print(BC)
    
```

```python
testout_df['Barcode'] = BC_list
testout_df
```
